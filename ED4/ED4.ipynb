#### Importação de bibliotecas e imagens

```python
# Imports
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import cv2 as cv
import skimage

def loadImage():
    img = cv.imread("lena.pgm")
    imgUint = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
    imgDouble = skimage.img_as_float(imgUint)
    return imgUint, imgDouble

imgUint, imgDouble = loadImage()
imgs = plt.subplots(1, 2, figsize=(10, 5))
imgs[1][0].imshow(imgUint, cmap="gray")
imgs[1][0].set_title("uint8")
imgs[1][1].imshow(imgDouble, cmap="gray")
imgs[1][1].set_title("double")
imgUint.dtype, imgDouble.dtype
```

#### Brightness

```python
imgUintBright, imgDoubleBright = loadImage()

#brightness on lena int and double
gain = 0.4 # Ganho em %
gainToUint = gain*255


imgUintBright = cv.add(imgUintBright, gainToUint)
imgDoubleBright = cv.add(imgDoubleBright, gain)
imgDoubleBright[imgDoubleBright > 1] = 1

cv.imwrite("lena_brightness.png", imgUintBright)

imgs = plt.subplots(1, 2, figsize=(10, 5))
imgs[1][0].imshow(imgUintBright, cmap="gray")
imgs[1][0].set_title("uint8 | +%.i brightness" % gainToUint)
imgs[1][1].imshow(imgDoubleBright, cmap="gray")
imgs[1][1].set_title("double | +%.2f brightness" % gain)
```

#### Constrast

```python
imgUintContrast, imgDoubleConstrast = loadImage()

#brightness on lena int and double
gain = 0.05 # Ganho em %


imgUintContrast = cv.multiply(imgUintContrast, gain)
imgDoubleConstrast = cv.multiply(imgDoubleConstrast, gain)
imgDoubleConstrast[imgDoubleConstrast > 1] = 1

cv.imwrite("lena_constrast.png", imgUintContrast)

imgs = plt.subplots(1, 2, figsize=(10, 5))
imgs[1][0].imshow(imgUintContrast, cmap="gray")
imgs[1][0].set_title("uint8 | %.2f constrast" % gain)
imgs[1][1].imshow(imgDoubleConstrast, cmap="gray")
imgs[1][1].set_title("double | %.2f constrast" % gain)
```

#### Negative

```python
imgUintNegative, imgDoubleNegative = loadImage()

#brightness on lena int and double
gain = 0.05 # Ganho em %


imgUintNegative = 255 - imgUintNegative
imgDoubleNegative = 1 - imgDoubleNegative

cv.imwrite("lena_negative.png", imgUintNegative)

imgs = plt.subplots(1, 2, figsize=(10, 5))
imgs[1][0].imshow(imgUintNegative, cmap="gray")
imgs[1][0].set_title("uint8 | negative")
imgs[1][1].imshow(imgDoubleNegative, cmap="gray")
imgs[1][1].set_title("double | negative")
```

#### Posterisation

```python
imgUintPosterisation, imgDoublePosterisation = loadImage()

#brightness on lena int and double
gain = 0.20 # Ganho em %
uIntgain = gain * 255


imgUintPosterisation = imgUintPosterisation // uIntgain * uIntgain
imgDoublePosterisation = imgDoublePosterisation // gain * gain

cv.imwrite("lena_posterisation.png", imgUintPosterisation)

imgs = plt.subplots(1, 2, figsize=(10, 5))
imgs[1][0].imshow(imgUintPosterisation, cmap="gray")
imgs[1][0].set_title("uint8 | %.2f Posterisation" % uIntgain)
imgs[1][1].imshow(imgDoublePosterisation, cmap="gray")
imgs[1][1].set_title("double | %.2f Posterisation" % gain)
```
