```python
# Imports
import matplotlib.pyplot as plt
from PIL import Image, ImageDraw
import numpy as np
import cv2 as cv
import skimage
import os
```

```python
penguin = cv.imread("penguins.png")
penguin = cv.cvtColor(penguin, cv.COLOR_BGR2GRAY)
plt.imshow(penguin, cmap="gray")
```

```python
Kv = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])
Ku = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])

Iv = cv.filter2D(penguin, -1, Kv)
Iu = cv.filter2D(penguin, -1, Ku)
Iuv = cv.filter2D(penguin, -1, Kv + Ku)
cv.imwrite("penguin_Iv.png", Iv)
cv.imwrite("penguin_Iu.png", Iu)
cv.imwrite("penguin_Iuv.png", Iuv)
edges = plt.subplots(1, 3, figsize=(12, 5))
edges[1][0].imshow(Iv, cmap="gray")
edges[1][0].set_title("Vertical edges")
edges[1][1].imshow(Iu, cmap="gray")
edges[1][1].set_title("Horizontal edges")
edges[1][2].imshow(Iuv, cmap="gray")
edges[1][2].set_title("Vertical and horizontal edges")
```

```python
I = cv.addWeighted(Iu, 0.5, Iv, 0.5, 0)
cv.imwrite("penguin_I.png", I)
plt.imshow(I, cmap="gray")
```
