#### Importação de bibliotecas e imagens

```python
# Imports
import matplotlib.pyplot as plt
import numpy as np
import cv2 as cv
import skimage

img = cv.imread("flowers4.png")
img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
img = skimage.img_as_float(img)
plt.imshow(img, cmap="gray")
img.dtype
```

```python
kU = np.ones((15, 15)) / 225
plt.imshow(kU, cmap="gray")
```

```python
imU = cv.filter2D(img, -1, kU)
plt.imshow(imU, cmap="gray")
```

```python

 # Largura do kernel definida como 17 para que seja o dobro da meia largura e impar
kG = cv.getGaussianKernel(17, 5)
plt.imshow(kG, cmap="gray")
```

```python
imG = cv.filter2D(img, -1, kG)
plt.imshow(imG, cmap="gray")
```

```python
comp= plt.subplots(1, 3, figsize=(15, 5))
comp[1][0].imshow(img, cmap="gray")
comp[1][0].axis('off')
comp[1][0].set_title("Original")
comp[1][1].imshow(imU, cmap="gray")
comp[1][1].axis('off')
comp[1][1].set_title("imU")
comp[1][2].imshow(imG, cmap="gray")
comp[1][2].axis('off')
comp[1][2].set_title("imG")
```

```python
cv.imwrite("flowers4_imU.png", imU * 255)
cv.imwrite("flowers4_imG.png", imG * 255)
```
