#### Importações

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2 as cv
from copy import copy

img = cv.imread('church.png')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img)
plt.axis('off')
plt.show()

```

```python
# Converter para escala de cinza
gray = cv.cvtColor(img, cv.COLOR_RGB2GRAY)

bordas = cv.Canny(gray, 50, 150, apertureSize=3)

# Aplicando a transformada de Hough
linhas = cv.HoughLinesP(bordas, 1, np.pi/180, threshold=100, minLineLength=100, maxLineGap=7)

result = copy(img)
height, width, _ = result.shape
black_canvas = np.zeros((height, width, 3), dtype=np.uint8)


# Desenhar as linhas dectadas
for linha in linhas:
    x1, y1, x2, y2 = linha[0]
    cv.line(result, (x1, y1), (x2, y2), (255, 0,0), 2)
    cv.line(black_canvas, (x1, y1), (x2, y2), (0, 255, 0), 2)

    
plt.imshow(img)
plt.axis('off')
plt.show()


plt.imshow(result)
plt.axis('off')
plt.show()

plt.imshow(black_canvas)
plt.axis('off')
plt.show()
```
