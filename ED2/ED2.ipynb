```python
# Imports
import matplotlib.pyplot as plt
from PIL import Image, ImageDraw
import numpy as np
import cv2 as cv
import skimage
import os
```

```python
image = cv.imread("castle.jpg")
image = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
plt.imshow(image, cmap="gray")
```

```python
H = cv.calcHist([image], [0], None, [256], [0, 256])
plt.plot(H, color="black")
plt.xlabel('Greylevel')
plt.ylabel('Number of samples')
```

```python
M = H.argmax()
print("M =", M)
T = 180 # Valor determinado visualmente pelo vale entre os picos
print("T =", T)


plt.plot(H, color="black", label="Data")
plt.axvline(x=M, color='blue', linestyle='--', label='Máximo')
plt.axvline(x=T, color='red', linestyle='--', label='Threshhold')
plt.xlabel('Greylevel')
plt.ylabel('Number of samples')
plt.show()
```

```python
#thresholding
ret, thresh = cv.threshold(image, T, 255, cv.THRESH_BINARY)
cv.imwrite("binaryImage.png", thresh)
plt.imshow(thresh, cmap="gray")
```
